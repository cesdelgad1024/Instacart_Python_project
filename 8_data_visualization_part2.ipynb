### Part 1:

```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
#import the orders_products_combined
df_merged = pd.read_pickle(r'C:\Users\Sthif\01-2025-Instacart Basket Analysis\02 Data\Prepared Data\instacart_data.pkl')
path = r'C:\Users\Sthif\01-2025-Instacart Basket Analysis'
df_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'instacart_data.pkl'))
```

### Part 3:

```python
# Histogram of 'order_hour_of_day'
hist = df_merged['order_hour_of_day'].plot(kind='hist', bins=24, edgecolor='black', color='skyblue')
plt.title('Orders by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')
plt.xticks(range(0, 24))  # Ensure all hours (0-23) are labeled
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional gridlines for better readability
plt.show()
```

##### Orders by Hour of the Day
###### The histogram demonstrates the distribution of orders placed at different hours of the day.

###### Key observations:

###### - Most orders are placed between 10 AM and 3 PM, with a noticeable peak around 12 PM (noon).
###### - Early morning hours (midnight to 6 AM) show significantly fewer orders, which indicates less customer activity during these times.
###### - The data helps identify peak operational times when more resources might be needed to handle demand efficiently.

```python
# Save the chart as a PNG file
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'histogram_order_hourly.png'))
```

### Part 4:

```python
# Create a bar chart for 'loyalty_flag'
fig, ax = plt.subplots(figsize=(8, 6))  # Create a figure and axis object
bar = df_merged['loyalty_flag'].value_counts().plot(kind='bar', color='lightcoral', edgecolor='black', ax=ax)
ax.set_title('Distribution of Orders by Loyalty')
ax.set_xlabel('Loyalty Flag')
ax.set_ylabel('Number of Orders')
ax.set_xticks(range(len(df_merged['loyalty_flag'].value_counts())))
ax.set_xticklabels(df_merged['loyalty_flag'].value_counts().index, rotation=0)
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Save the chart as a PNG file
bar.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'dis_orders_loyalty.png'))
```

### Step 5:

```python
# Take a random sample of 100,000 rows for better performance
sample_df = df_merged.sample(n=100000, random_state=42)
```

```python
# Group by 'order_hour_of_day' and calculate mean prices
hourly_prices = sample_df.groupby('order_hour_of_day')['prices'].mean()
```

```python
# Create the line chart
plt.figure(figsize=(10, 6))
line = hourly_prices.plot(kind='line', marker='o', color='green', linestyle='-')
plt.title('Average Expenditure by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Price ($)')
plt.xticks(range(0, 24))  # Ensure all hours (0-23) are labeled
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines for readability
plt.show()
```

##### Average Expenditure by Hour of the Day
###### This line chart illustrates the average expenditure (prices) across different hours of the day. Peaks in the chart indicate hours where 
###### customers tend to spend more, while troughs indicate hours with lower average expenditures. The data can help identify when customers
###### are likely to purchase higher-value items and assist in demand planning.

```python
# Save the chart as a PNG file
line.figure.savefig(r'C:\Users\Sthif\01-2025-Instacart Basket Analysis\04 Analysis\Visualizations\line_chart_hourly_prices.png')
```

### Step 6:

```python
# Group by 'age' and calculate the mean number of dependents
age_dependents = df_merged.groupby('age')['num_dependants'].mean()
```

```python
# Create the line chart
plt.figure(figsize=(10, 6))
line2 = age_dependents.plot(kind='line', marker='o', color='blue', linestyle='-')
plt.title('Average Number of Dependents by Age')
plt.xlabel('Age')
plt.ylabel('Average Number of Dependents')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Save the chart as a PNG file
line2.figure.savefig(r'C:\Users\Sthif\01-2025-Instacart Basket Analysis\04 Analysis\Visualizations\avg_num_dependets_by_age.png')
```
