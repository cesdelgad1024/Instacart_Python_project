```python
pd.merge(orders_products_combined , df_prods , on = ['product_id'] , how = 'inner', indicator = 'match' )
```

```python
# Dropping the "unnamed:0" column
df_prods = df_prods.drop(columns=['Unnamed: 0'])
```

```python
# Merge of 'orders_products_combined' and 'df_prods'
ords_prods_merge = orders_products_combined.merge(df_prods , on = ['product_id'] , indicator = 'match' )
```

```python
#Export the newly created dataframe as ords_prods_merge in a suitable format
# Dropping the "_merge and match" column
ords_prods_merge = ords_prods_merge.drop(columns=['_merge', 'match'])
```

```python
# Export Data in Pickle Format
ords_prods_merge.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_combined.pkl'))
```
